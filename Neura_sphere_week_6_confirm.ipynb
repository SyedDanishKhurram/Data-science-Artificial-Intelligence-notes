# Data Cleaning:

* Python Libraries: Python libraries like Pandas, NumPy, and OpenRefine can be used for data cleaning and transformation.



```python
import pandas as pd

# Sample dataset with duplicates and missing values
data = {
    'Name': ['John', 'Alice', 'Bob', 'John', 'Eve'],
    'Age': [25, None, 30, 25, 22],
    'Gender': ['Male', 'Female', 'M', 'Male', 'F']
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset:")
print(df)

# Data Cleaning Steps:

# 1. Remove Duplicates
df.drop_duplicates(inplace=True)

# 2. Handling Missing Values
df['Age'].fillna(df['Age'].median(), inplace=True)

# 3 Replace the entries
df['Gender'] = df['Gender'].replace(['M'],'Male')
df['Gender'] = df['Gender'].replace(['F'], 'Female')

# Display the cleaned dataset
print("\nCleaned Dataset:")
print(df)
#\n is an escape sequence that represents a newline character and breaks the line

```

# Data Validations:

* SQL Constraints: In the database, you can define constraints such as UNIQUE, NOT NULL, and CHECK constraints to enforce data validity.
Custom Validation Scripts: You can write custom validation scripts in Python, Java, or other programming languages to validate data before ingestion.

```python
data = {
    'Name': ['John', 'Alice', 'Bob', 'Eve', 'Chris'],
    'Age': [25, 30, 22, -5, 40],
    'Gender': ['Male', 'Female', 'M', 'F', 'Male'],
    'Salary': [50000, 60000, 45000, 55000, 700000]
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset:")
print(df)

# 3 Replace the entries
df['Gender'] = df['Gender'].replace(['M'],'Male')
df['Gender'] = df['Gender'].replace(['F'], 'Female')

# 1. Validate Age Values (Positive Values Only)
df = df[df['Age'] > 0]

# 2. Validate Salary Values (Realistic Range)
df = df[(df['Salary'] >= 30000) & (df['Salary'] <= 200000)]

# Display the cleaned and validated dataset
print("\nValidated Dataset:")
print(df)
```

```python
from datetime import datetime, timedelta
from airflow import DAG
from airflow.operators.python_operator import PythonOperator
from airflow.operators.email import EmailOperator

# Default arguments for the DAG
default_args = {
    'owner': 'airflow',
    'depends_on_past': False,
    'start_date': datetime(2023, 1, 1),
    'email_on_failure': False,
    'email_on_retry': False,
    'retries': 1,
    'retry_delay': timedelta(minutes=5),
}

# Create a DAG
dag = DAG(
    'example_dag',
    default_args=default_args,
    schedule_interval=timedelta(days=1),  # Run daily
)

# Define Python functions
def task1():
    print("Executing Task 1")

def task2():
    print("Executing Task 2")

# Create PythonOperator tasks
task1 = PythonOperator(
    task_id='task_1',
    python_callable=task1,
    dag=dag,
)

task2 = PythonOperator(
    task_id='task_2',
    python_callable=task2,
    dag=dag,
)

# Define email task
email = EmailOperator(
    task_id='send_email',
    to='youremail@example.com',
    subject='Airflow DAG Execution Report',
    html_content='<p>The Airflow DAG has completed successfully.</p>',
    dag=dag,
)

# Set task dependencies
task1 >> task2 >> email
```

```python
# Import necessary libraries
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
import matplotlib.pyplot as plt

# Initialize a SparkSession
spark = SparkSession.builder.appName("DatabricksExample").getOrCreate()

# Load a sample dataset (you can replace this with your own dataset)
data = spark.read.csv("give me a file for csv", header=True, inferSchema=True)

# Show the first few rows of the dataset
data.show()

# Data Exploration
# Calculate summary statistics
summary = data.describe()
summary.show()

# Perform data aggregation and analysis
agg_data = data.groupBy("Year").agg(F.avg("Price").alias("Avg_Price"), F.sum("Population").alias("Total_Population"))
agg_data.show()

# Data Visualization
# Convert the aggregated data to a Pandas DataFrame for visualization
agg_df = agg_data.toPandas()

# Create a line plot of average price over the years
plt.figure(figsize=(10, 6))
plt.plot(agg_df["Year"], agg_df["Avg_Price"], marker='o', linestyle='-')
plt.title('Average Price Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()

```
